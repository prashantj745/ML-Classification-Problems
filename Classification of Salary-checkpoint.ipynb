```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,f1_score

```

```python
## Import Dataset

df = pd.read_csv('C:/Users/Savita bhabhi/Desktop/ML/Salary Classification/adult_test.csv')
df.head()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df = df[(df.astype(str) != ' ?').all(axis=1)]
df.shape


income = df['income']
X = df.drop('income',axis =1)
#X = df.iloc[:,:-1]
X.shape
income.shape
```

```python
print(X.head())
```

```python
income_final = income.map({' <=50K.':0,' >50K.':1})
#income_final =  income.apply(lambda x: 1 if x == '>50K' else 0)
#df.apply(lambda x : x['Salary'] == '>=50k' )]
income_final.head(50)
```

```python
X.shape
income.shape


skewed = ['capital-gain', 'capital-loss']
X_log_transformed = pd.DataFrame(data = X)
X_log_transformed[skewed] = X[skewed].apply(lambda x: np.log(x + 1))

X.skew()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
columns_to_norm = ['Age', 'fnlwgt', 'education-num','capital-gain', 'capital-loss', 'hours-per-week']
X_norm_data = pd.DataFrame(data = X_log_transformed)
X_norm_data[columns_to_norm] = scaler.fit_transform(X_log_transformed[columns_to_norm])
X_norm_data.head()
```

```python
#One hot encoding 

from sklearn.preprocessing import OneHotEncoder

X_encoded = pd.get_dummies(X_norm_data)
#X_encoded = X_encoded.drop(['Workclass_?'], axis=1,inplace =True)
X_encoded.head(10)
```

```python
# Split and train LR
X_train,X_test,y_train,y_test = train_test_split(X_encoded,income_final,test_size = 0.3,random_state = 42)

#Record run time
import time 
start_time = time.time()

# Apply LogisticRegression

clf_lr = LogisticRegression()
clf_lr.fit(X_train,y_train)
end_time = time.time()

print(str(end_time - start_time)+ " seconds")

y_pred = clf_lr.predict(X_test)

score = clf_lr.score(X_test, y_test)
print("Accuracy Score :" + str(score*100))

from sklearn import metrics

cf_clf_lr  = metrics.confusion_matrix(y_test,y_pred)

from sklearn.metrics import f1_score

f1_clf_lr  = metrics.f1_score(y_test,y_pred)

clas_clf_lr  = metrics.classification_report(y_test,y_pred)
print ("Confusion matrix : " + str(cf_clf_lr))
print("F1_score : " + str(f1_clf_lr))
print("Classification report " +  str(clas_clf_lr))

print (X_test.shape)
```

```python
# Function to define Pipeline 


def pipeline(clf_name, X_train,X_test,y_train,y_test):
    
    start_time = time.time()
    clf_name.fit(X_train,y_train)
    end_time = time.time()
    clf_time = end_time - start_time
    print("Time taken by {} :{} seconds ".format((clf_name), (clf_time)))
    
    y_pred = clf_name.predict(X_test)
    
        
    #Accuracy score
    from sklearn.metrics import accuracy_score
    
    clf_acc = accuracy_score(y_test,y_pred)
    
    print( "Accuracy Score for {} : {}".format(clf_name , clf_acc))
    
   
    
    
    
    
    
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

clf_lr = LogisticRegression()
clf_KNN = KNeighborsClassifier(n_neighbors=5,n_jobs = -1)
clf_NB = GaussianNB()
clf_DT = DecisionTreeClassifier(criterion='gini', max_depth=5)
clf_SVM = SVC(random_state=0)
results = {}

for clf in [ clf_lr, clf_KNN, clf_NB]:
    results[clf] = pipeline(clf, X_train,X_test,y_train,y_test)
    print(results)
    

    
```
